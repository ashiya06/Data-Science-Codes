```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.histplot(df['Fare'])
plt.show()
```

```python
sb.histplot(df['Fare'],color="olive", binwidth=30)
plt.show()
```

```python
#sb.histplot(df['Fare'],kde=True,color="olive", binwidth=30)
sb.histplot(data=df, x="Fare", hue="Pclass", binwidth=30)
plt.show()
```

```python
sb.histplot(df['Fare'],kde=True,color="olive", binwidth=30)
plt.show()
```

```python
import seaborn as sb
sb.distplot(df['Fare'])
plt.show()
```
